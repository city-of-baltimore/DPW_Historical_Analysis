### Defined Functions 

```python
def due_date(create_date,sla):
    due = dt.datetime.strptime(create_date, '%Y-%m-%d').date() + dt.timedelta(days=sla)
    return str(due)
def lower_str(text):
    return text.lower()
def textstr2date(text):
    if pd.notnull(text):
        new_str=str(text)[0:10]
    else:
        new_str = text
    return new_str
def oopen(data_inp,date_comp,filter_old = False,filter_old_years = 2):

    #earliest_create = pd.to_datetime(a.strftime('%Y-%m-%d')) - pd.to_timedelta(filter_old_years, 'Y')
    #COUNTS NUMBER OPEN
    count_close=len(data_inp[(data_inp['Date_Create']<date_comp)&((data_inp['Status'].apply(lower_str).str.contains('close'))&(data_inp['Date_Finish']>date_comp))])
    count_cancel=len(data_inp[(data_inp['Date_Create']<date_comp)&((data_inp['Status'].apply(lower_str).str.contains('cancel'))&(data_inp['Date_Finish']>date_comp))])
    count_new =len(data_inp[(data_inp['Date_Create']<date_comp)&(data_inp['Status'].apply(lower_str).str.contains('new'))])
    count_pending =len(data_inp[(data_inp['Date_Create']<date_comp)&(data_inp['Status'].apply(lower_str).str.contains('pending'))])
    counter=count_close+count_new+count_cancel
    #Doesn't count PENDING
    return counter

def oover(data_inp,date_comp):
    earliest_create = pd.to_datetime(a.strftime('%Y-%m-%d')) - pd.to_timedelta(2, 'Y')
    #COUNTS NUMBER OPEN
    count_close=len(data_inp[(data_inp['Due_date']<date_comp)&((data_inp['Status'].apply(lower_str).str.contains('close'))&(data_inp['Date_Finish']>date_comp))])
    count_cancel=len(data_inp[(data_inp['Due_date']<date_comp)&((data_inp['Status'].apply(lower_str).str.contains('cancel'))&(data_inp['Date_Finish']>date_comp))])
    count_new =len(data_inp[(data_inp['Due_date']<date_comp)&(data_inp['Status'].apply(lower_str).str.contains('new'))])
    count_pending =len(data_inp[(data_inp['Due_date']<date_comp)&(data_inp['Status'].apply(lower_str).str.contains('pending'))])
    counter=count_close+count_new+count_cancel+count_cancel
    #Doesn't count PENDING
    return counter

def days_between(d1, d2):
    d1 = dt.datetime.strptime(d1, "%Y-%m-%d")
    d2 = dt.datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)
```

### Data Cleaning

#### Dropping NAs, reducing fields, renaming fields

```python
chip_wo_red = chip_data[['Record_ID','Date_Create','Date_Finish','quadrant','Priority','Clean_Size','Crew_Name','Status','VRI_Focus_Area','Work_Order_Type']]
chip_wo_red.columns = ['Record_id','Date_Create','Date_Finish','Quadrant','Priority','Clean_Size','Crew','Status','VRI_Focus_Area','Type']
cw_wo_red = cw_wo[['Record_ID','Actual_Start_Date','Actual_Finish_Date','Shop','Priority','Category_Size','Work_Completed_By','Status','VRI_Focus_Area']]
cw_wo_red = cw_wo_red.assign(Category_Size='ProMow')
cw_wo_red['Type'] = 'ProMow'
cw_wo_red.columns = ['Record_id','Date_Create','Date_Finish','Quadrant','Priority','Clean_Size','Crew','Status','VRI_Focus_Area','Type']
```

#### Aggregating Datasets, drop NAs

```python
all_wo_comb = pd.concat([chip_wo_red,cw_wo_red],ignore_index=True)
all_wo_comb = all_wo_comb.drop(all_wo_comb[all_wo_comb['Quadrant'].isna()].index)
all_wo_comb = all_wo_comb.drop(all_wo_comb[all_wo_comb['Status'].isna()].index)
all_wo_comb = all_wo_comb.drop(all_wo_comb[all_wo_comb['Date_Create'].isna()].index)
#all_wo_comb = all_wo_comb.drop(all_wo_comb[all_wo_comb['Date_Create']>'2019-12-31'].index)
all_wo_comb['Quadrant'] = all_wo_comb['Quadrant'].replace(['NW','NE','SW','SE','Downtown'],['NORTHWEST','NORTHEAST','SOUTHWEST','SOUTHEAST','CENTRAL'])
all_wo_comb['VRI_Focus_Area'] = all_wo_comb['VRI_Focus_Area'].fillna('Other')
all_wo_comb = all_wo_comb.drop(all_wo_comb[all_wo_comb['Quadrant']=='CENTRAL'].index)
all_wo_comb['Date_Create'] = all_wo_comb['Date_Create'].apply(textstr2date)
all_wo_comb['Date_Finish'] = all_wo_comb['Date_Finish'].apply(textstr2date)
all_wo_comb['Year'] = pd.DatetimeIndex(all_wo_comb['Date_Finish']).year
all_wo_comb['Month'] = pd.DatetimeIndex(all_wo_comb['Date_Finish']).month
all_wo_comb['Date'] = pd.to_datetime(all_wo_comb[['Year', 'Month']].assign(DAY=1))
all_wo_comb['Date'] = all_wo_comb['Date'].dt.date
all_wo_comb.loc[:,'Due_date'] = all_wo_comb["Date_Create"].apply(due_date,args=(42,))
all_wo_comb = all_wo_comb.reset_index()
```

#### Date Between WorkOrders opened and closed by type

```python
date_diff_query='''
        SELECT [Record ID] as id, 
        cast([Date Create] as DATE) as open_date, 
        case when [Date Finish] is null then cast('2019-12-31' as DATE)
            when [Date Finish] is not null then cast([Date Finish] as DATE) 
            as close_date,
        [Clean Size] as clean_size, 
        DATEDIFF(day, [Date Create],close_date) as date_difference
        FROM CitiStat.dbo.CHIP_WorkOrders
        WHERE 1=1
        AND [Work Order Type] = 'Cleaning'
        AND [Date Create] > '2019-04-01'
        order by clean_size
'''
date_diff=pd.read_sql(date_diff_query, conn,index_col = 'id')
date_diff.head()
```
